```python
import pandas as pd
#open the CSV file
df = pd.read_csv('fish.csv') 
df
```

```python
#the models
y = df.loc[:,['Weight']]
y
```

```python
#the features
x = df.loc[:,['Species','Length1','Length2','Length3','Height','Width']]
x
```

```python
#check for the types fish listed in the dataset
x['Species'].value_counts()
```

```python
x['Species'] = x['Species'].replace({'Bream' : 1, 'Roach' : 2, 'Whitefish' : 3, 'Parkki' : 4, 'Perch' : 5, 'Pike' : 6, 'Smelt' : 7})
x
#if we want to return them back, we use the following:
# x['Species'] = x['Species'].replace({'Bream' : 1, 'Roach' : 2, 'Whitefish' : 3, 'Parkki' : 4, 'Perch' : 5, 'Pike' : 6, 'Smelt' : 7})
```

```python
#just a check to see if the strings were replaced
x.iloc[54:64]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 25)
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
#converting the array into a dataframe for better visualization 
Y_predicted = pd.DataFrame(y_pred)
Y_predicted
```

```python
y_test
```

```python
from sklearn import metrics
import numpy as np
# Mean squared error: metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
#we can't have a percentage because this is a linear regression model not a classification model
```

We can see that the RMSE is a bit high, which means our model is no that accurate, and the reason for this could be
the small database we dealt with.

I have tested the model using different features and turned out that using the mean of the 3 Lengths was not the best choice,
therefore I used all three columns of Length to get the best accuracy (lowest error rate). 
The RMSE for the model using the Lengths' mean was 125.69 ,
whereas using all Length columns the RMSE was 125.47
